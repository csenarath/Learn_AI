```python
from langchain.messages import HumanMessage, AIMessage, SystemMessage
from langchain_anthropic import ChatAnthropic
```

```python
import os
from dotenv import load_dotenv
from langchain_anthropic import ChatAnthropic

# Load environment variables from .env file
load_dotenv()

# Get API key from environment
API_KEY = os.getenv("API_KEY")

# Create LangChain Claude client
llm_anthropic = ChatAnthropic(
    model="claude-sonnet-4-5",
    api_key=API_KEY,
    max_tokens=1024
)

from langchain_core.prompts import PromptTemplate ,ChatPromptTemplate
```

```python
llm_anthropic.invoke("Tell me a Joke")
```

```python
instruction = 'generate output in key-value pair format with setup and punchline'

llm_anthropic.invoke(f"Tell me a Joke. {instruction}")
```

Pydentic


```python
from pydantic import BaseModel ,Field

class llm_schema(BaseModel):
    setup: str = Field(discription="the setup for the joke")
    punchline: str  = Field(discription="the puchline for the joke")
```

```python
obj1=llm_schema(**{"setup":"some setup", "punchline":"some puchline"})
```

```python
llm_structured_output= llm_anthropic.with_structured_output(llm_schema)
```

```python
result=llm_structured_output.invoke("tell me a joke")
```

```python
result.punchline
```
